# Plots standalone colorbars for the Tb + precipitation animations

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
import colorcet as cc
```

```python
PHASE = 'Summer'
fig_dir = f'/global/cfs/cdirs/m1867/zfeng/MCSMIP/DYAMOND/{PHASE}/quicklooks/animations/'
fig_dir
```

```python
# Colormap transparency
alpha_tb = 0.9
alpha_pcp = 0.3

# Specify colormaps
tb_cmap = cc.cm['gray_r']
pcp_cmap = cc.cm["CET_R1"]

# Tb and precipitation color levels
pcp_levels = [2, 3, 4, 5, 6, 8, 10, 15, 20, 30]
tb_levels = np.arange(180, 300.1, 1)

tb_ticks = np.arange(180, 300.1, 10)
pcp_ticks = pcp_levels

tb_label = 'Tb (K)'
pcp_label = 'Precipitation (mm h$^{-1}$)'
```

```python
mpl.rcParams['font.size'] = 30
mpl.rcParams['font.family'] = 'Helvetica'
```

```python
cmap = plt.get_cmap(tb_cmap)
# cmap.set_under()
norm = mpl.colors.BoundaryNorm(tb_levels, ncolors=cmap.N, clip=True)

# Set the alpha for the colormap (transparency)
cmap._init()  # Initialize the colormap
cmap._lut[:-3, -1] = alpha_tb  # Set the alpha channel (last column) for all entries except the last 3

# Create a figure and a single subplot
fig, ax = plt.subplots(figsize=(20, 0.5), dpi=100)

# Create a colorbar in the subplot
cbar = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), cax=ax, orientation='horizontal', extend='both', ticks=tb_ticks, label=tb_label)

# # Set the alpha for each patch in the colorbar
# for patch in cbar.ax.patches:
#     patch.set_alpha(alpha_tb)

figname = f'{fig_dir}colorbar_tb_horizontal.png'
fig.savefig(figname, dpi=300, bbox_inches='tight', facecolor='w')
```

```python
cmap = plt.get_cmap(pcp_cmap)
# cmap.set_under()
norm = mpl.colors.BoundaryNorm(pcp_levels, ncolors=cmap.N, clip=True)

# Set the alpha for the colormap (transparency)
cmap._init()  # Initialize the colormap
cmap._lut[:-3, -1] = alpha_pcp  # Set the alpha channel (last column) for all entries except the last 3

# Create a figure and a single subplot
fig, ax = plt.subplots(figsize=(20, 0.5), dpi=100)

# Create a colorbar in the subplot
cbar = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap), cax=ax, orientation='horizontal', extend='both', ticks=pcp_ticks, label=pcp_label)

# # Set the alpha for each patch in the colorbar
# for patch in cbar.ax.patches:
#     patch.set_alpha(alpha_pcp)

figname = f'{fig_dir}colorbar_pcp_horizontal.png'
fig.savefig(figname, dpi=300, bbox_inches='tight', facecolor='w')
```
